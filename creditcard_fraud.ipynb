```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("creditcard.csv")
data
```

```python
data=data.dropna()
```

```python
data.columns
```

```python
fraud_trans=data[data["Class"]==1]
valid_trans=data[data["Class"]==0]
outlier_frac=len(fraud_trans)/len(valid_trans)
print(outlier_frac)
```

```python
print("fraud cases:{}".format(len(data[data["Class"]==1])))
print("valid cases:{}".format(len(data[data["Class"]==0])))
```

```python
data.columns
```

```python
fraud_trans.Amount.describe()
```

```python
valid_trans.Amount.describe()
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
best_result=[]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=0)
```

###DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
model1=DecisionTreeClassifier(random_state=0)
model1.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
y_pred=model1.predict(x_test)
print(confusion_matrix(y_test,y_pred))
ans=accuracy_score(y_test,y_pred)
best_result.append(ans)
ans
```

### K-NN

```python
from sklearn.neighbors import KNeighborsClassifier
model2=KNeighborsClassifier()
model2.fit(x_train,y_train)
```

##Confusion matrix of knn

```python
from sklearn.metrics import confusion_matrix,accuracy_score
y_pred=model2.predict(x_test)
print(confusion_matrix(y_test,y_pred))
ans=accuracy_score(y_test,y_pred)
best_result.append(ans)
ans
```

##kernel_svm

```python
from sklearn.svm import SVC
model3=SVC(random_state=0)
model3.fit(x_train,y_train)
```

##confusion matrix of kernel_svm

```python
from sklearn.metrics import confusion_matrix,accuracy_score
y_pred=model3.predict(x_test)
print(confusion_matrix(y_test,y_pred))
ans=accuracy_score(y_test,y_pred)
best_result.append(ans)
ans
```

##Logistic_regression

```python
from sklearn.linear_model import LogisticRegression
model4=LogisticRegression(random_state=0)
model4.fit(x_train,y_train)
```

##confusion matrix of logistic_regression

```python
from sklearn.metrics import confusion_matrix,accuracy_score
y_pred=model4.predict(x_test)
print(confusion_matrix(y_test,y_pred))
ans=accuracy_score(y_test,y_pred)
best_result.append(ans)
ans
```
